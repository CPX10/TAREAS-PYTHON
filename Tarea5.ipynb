# TECHNO IMPORT
IMPORTADORES DE EQUIPOS COMPUTACIONALES

```python
print("Ingrese la medida de su paquete en cm: ")
ancho = float(input("Ancho en cm: "))
alto = float(input("Alto en cm: "))
largo = float(input("Largo en cm: "))
pesoKg = float(input("peso en Kg: "))
```

```python
# peso volumen Kg= (Largo * alto * ancho)/5000

pesoVolumen = (largo*ancho*alto)/5000
precioKg = 6.50 #usd precio que cobra la empresa por enviar cada kilogramo
impuesto = 0.15 #porcentaje de impuesto
```

```python
pesoVolumen
```

```python
pesoKg
```

```python
resultadoCosto = precioKg * pesoKg

if pesoVolumen > pesoKg:
    resultadoCosto = precioKg * pesoVolumen

valorImpuesto = resultadoCosto * impuesto
resultadoCosto
```

```python
precioKg * pesoVolumen
```

```python
precioKg * pesoKg
```

# RESULTADO COSTOS


```python
print(f"Precio a pagar: USD $.{resultadoCosto}" )
print("impuesto del {}% USD $.{}".format(int(impuesto*100),valorImpuesto))
print("TOTAL DE USD $" + str(valorImpuesto + resultadoCosto))
```
